```python
import pandas
import numpy as np
exam_dist = np.random.normal(loc=65,scale=6,size=(8))
exam_dist
```

```python
labels = []
for i in range(0,8):
    code = "CS1"+str(np.random.randint(10,40))
    while code in labels:
        code = "CS1"+str(np.random.randint(10,40))
    labels.append(code)
exam_series = pandas.Series(exam_dist,labels)
exam_series
```

```python
students = []
for i in range(0,100):
    stu_id = "220"+str(np.random.randint(1000,9999))
    while stu_id in students:
        stu_id = "220"+str(np.random.randint(1000,9999))
    students.append(stu_id)
students
```

```python
data = []
for student in students:
    print("Student: "+ student)
    stu_dev = np.random.normal(loc=65,scale=10)
    print("\tStudent deviation: "+str(stu_dev))
    marks = []
    for exam in exam_series:
        exam_mean = (exam * stu_dev)/65
        marks.append(np.round(np.random.normal(loc=exam_mean,scale=6,size=(1))[0],2))
    data.append(marks)
data
```

```python
mark_data = pandas.DataFrame(data,columns=labels, index=students)
mark_data.index.name = "Student"
mark_data
```

```python
mark_data.to_csv('.\\Data\\student_data2.csv',index=True)
```
